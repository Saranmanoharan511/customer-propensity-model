```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Saranmanoharan511/customer-propensity-model/main/customer-propensity-model/online_retail.csv.zip", compression="zip")
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["CustomerID"])
```

```python
df.isnull().sum()
```

```python
df=df[df["Quantity"]>0]
df=df[df["UnitPrice"]>0]
```

```python
df.shape
```

```python
df["InvoiceDate"].dtype
```

```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
```

```python
df.info()
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

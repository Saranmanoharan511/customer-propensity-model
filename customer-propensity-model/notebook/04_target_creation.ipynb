```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Saranmanoharan511/customer-propensity-model/main/customer-propensity-model/cleaned_data.zip", compression="zip")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
rfm=pd.read_csv("https://raw.githubusercontent.com/Saranmanoharan511/customer-propensity-model/main/customer-propensity-model/rfm_features.zip", compression="zip")
```

```python
rfm
```

```python
cutoff_date=df['InvoiceDate'].max()-pd.Timedelta(days=30)
```

```python
cutoff_date
```

```python
train_df=df[df['InvoiceDate']<=cutoff_date]
future_df=df[df['InvoiceDate']>cutoff_date]
```

```python
future_df
```

```python
target=future_df.groupby("CustomerID").size()>0
target=target.astype(int)
```

```python
target
```

```python
rfm["Target"]=rfm["CustomerID"].map(target)
rfm["Target"]=rfm['Target'].fillna(0)
```
